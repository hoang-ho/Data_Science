#### Unsupervised and Supervised Learning based on likes, dislikes, views, comment_counts, and trend_duration

```python
dnew = df2[['category', 'trend_duration', 'publish_to_trend', 'publish to last trending', 'views', 'likes', 'dislikes', 'comment_count']]
dnew['multi_trend'] = (dnew['trend_duration'] != 0).astype(int)
dnew
```

```python
X = dnew[['publish_to_trend', 'views', 'likes', 'dislikes', 'comment_count']].as_matrix()
y = dnew['category']
```

```python
from sklearn.cluster import KMeans
```

Let's see if k-means can find any pattern among different categories

```python
kmeans = KMeans(n_clusters = y.nunique() , random_state = 0).fit(X)
predict = kmeans.predict(X)
print(pd.crosstab(predict,y,rownames = ['clusters']))
```

... What's about 2 clusters...

```python
kmeans = KMeans(n_clusters = 2, random_state = 0).fit(X)
predict = kmeans.predict(X)
print(pd.crosstab(predict,y,rownames = ['clusters']))
```

Still no relationship, almost everything ends up in the first cluster ... What's about pattern among videos that trend for multiple days and videos that don't

```python
y = dnew['multi_trend'].values
```

```python
kmeans = KMeans(n_clusters = 2 , random_state = 0).fit(X)
predict = kmeans.predict(X)
print(pd.crosstab(predict,y,rownames = ['clusters']))
```

As we can see, k-means cannot identify the difference between videos that trend for multiple days and videos that are from difference categories. Now, let's try some supervised learning algorithms

```python
from sklearn.cross_validation import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('accuracy ', accuracy)
```

Logistic Regression did well in indentifying videos that trend for several days and those that don't

```python
from sklearn import neighbors
```

```python
import itertools
```

```python
Metrics = ['euclidean', 'minkowski', 'cosine', 'jaccard']
K = [1,3,5,15,25,50]
Weights = ['uniform', 'distance']
Combo = list(itertools.product(K,Weights, Metrics))
len(Combo)
```

```python
for c in Combo:
    clf = neighbors.KNeighborsClassifier(c[0], weights = c[1], metric = c[2])
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print '# k neighbors: %i, weights: %s, metric: %s' % (c[0], c[1], c[2])
    print 'Test Perf ACC: %3f' % accuracy
```

K-NN also provides an equally good result

### Sentimental Analysis

```python
from wordcloud import WordCloud
import nltk

from nltk.sentiment import SentimentIntensityAnalyzer
from nltk.corpus import stopwords
from nltk import sent_tokenize, word_tokenize
from wordcloud import WordCloud, STOPWORDS

from collections import Counter
#from nltk.tokenize.regexp import RegexpTokenize
import re
```

Let's look at the tag words and find out the Polarity of categories on Youtube! It will take a while...

```python
polarities = list()
for cate in category_dict.values():
    tag_words = df2[df2['category'] == cate]['tags'].str.lower().str.cat(sep = ' ')
    tag_words = re.sub('[^A-Za-z]+', ' ', tag_words) #take everything that is characters
    tokens = word_tokenize(tag_words) #create a list of words from the string
    token_words = [w for w in tokens if w not in list(stopwords.words('english')) and len(w) > 2] #take everything that is not a stopword and has length bigger than 2
    word_dist = nltk.FreqDist(token_words) #get frequency of each word in token_words
    
    temp_df = pd.DataFrame(word_dist.most_common(1000), columns = ['Word', 'Frequency'])
    
    compound = 0
    for word in temp_df['Word']:
        compound += SentimentIntensityAnalyzer().polarity_scores(word)['compound'] #apply sentiment analysis to the word
    
    polarities.append(compound)
```

```python
sentiment_df = pd.DataFrame([category_dict.values(), polarities]).T
sentiment_df.columns = ['Category', 'Polarities']
sentiment_df.sort_values('Polarities', inplace = True)

sentiment_df
```

```python
plt.figure(figsize=(16,10))
sns.set(style="white",context="talk")
ax = sns.barplot(x=sentiment_df['Polarities'],y=sentiment_df['Category'], data=sentiment_df,orient='h',palette="RdBu")
plt.xlabel("Categories")
plt.ylabel("polarity")
plt.title("Polarity of Categories in Youtube videos")
```

By using sentiment analyzer from NLTK, we can examine the polarities of tags from all Youtube Trending Videos. All the 1000 most frequent tag from each category were examined and result in form of number. NEGATIVE values indicate that most of the tags have negative sentiment and POSITIVE in contrast. The result is as expected

```python
temp_df = temp_df.drop_duplicates('video_id', keep = 'last')
temp_df
```

```python
full_df = comments.merge(temp_df, how = 'left', left_on = 'video_id', right_on = 'video_id')
full_df
```

Now, let's attempt to predict the category of the youtube video again by using the comment text and see how it goes

```python
labels = []
all_texts = []
for x in full_df['video_id'].unique():
    temp_df = full_df[full_df['video_id'] == x]
    text = ""
    for c in temp_df['comment_text']:
        c = re.sub('[^A-Za-z]+', '', c)
        text += c + " "
    labels.append(temp_df['category'].values[0])
    all_texts.append(text)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(min_df=5)
features = vectorizer.fit_transform(all_texts)
```

```python
features.shape
```

```python
len(labels)
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.33, random_state = 42)
clf = LogisticRegression(solver = 'lbfgs', multi_class = 'multinomial')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print('accuracy ', accuracy)
```

```python
Metrics = ['euclidean', 'minkowski', 'cosine']
K = [1,3,5,15,25,50]
Weights = ['uniform', 'distance']
Combo = list(itertools.product(K,Weights, Metrics))
len(Combo)
```

```python
for c in Combo:
    clf = neighbors.KNeighborsClassifier(c[0], weights = c[1], metric = c[2])
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print '# k neighbors: %i, weights: %s, metric: %s' % (c[0], c[1], c[2])
    print 'Test Perf ACC: %3f' % accuracy
```

Logistic Regression does better this time than when using just likes, dislikes, views, comments count. Now it achieves a 30% accuracy. K-NN does a little bit better this time. Our best result is # k neighbors: 50, weights: distance, metric: cosine
Test Perf ACC: 0.332888
