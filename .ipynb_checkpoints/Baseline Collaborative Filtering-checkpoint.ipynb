```python
prediction_base, algo_base = dump.load("./output/output_BaselineOnly")
prediction_norm, algo_norm = dump.load("./output/output_NormalPredictor")
prediction_svd, algo_svd = dump.load("./output/output_SVD")
```

```python
df_svd = pd.DataFrame(prediction_svd, columns=['uid', 'bid', 'rui', 'est', 'details'])    
df_base = pd.DataFrame(prediction_base, columns=['uid', 'bid', 'rui', 'est', 'details'])  
df_norm = pd.DataFrame(prediction_norm, columns=['uid', 'bid', 'rui', 'est', 'details'])

df_svd['err'] = abs(df_svd.est - df_svd.rui)
df_base['err'] = abs(df_base.est - df_base.rui)
df_norm['err'] = abs(df_norm.est - df_norm.rui)
```
