# Cleaning Business

First let's just focus on food, drink and dessert businesses

```python
import pandas as pd
import re
import ast
import time
```

```python
business = pd.read_csv('business.csv')
```

```python
business.shape
```

```python
business['categories'].nunique()
```

```python
business.head()
```

```python
business['categories']
```

```python
business['categories'][0]
```

```python
ast.literal_eval(business.categories[0])
```

```python
start = time.time()
business["categories"] = [ast.literal_eval(business.categories[i]) for i in range(business.shape[0])]
end = time.time()
print(end-start)
```

```python
business["categories"]
```

```python
business.categories.values
```

```python
business.categories.values.tolist()
```

```python
category = {'Food', 'Restaurants', 'Bars', 'Pubs', 'Drink', 'Wine', 'Beer', 'Coffee', 'Tea', 'Bakeries', 'Bagel', 'Sandwich', 'Salad'}
```

```python
start = time.time()
business = business.loc[[i for i in range(business.shape[0]) if set(business.loc[i]['categories']) & category]]
end = time.time()
print(end - start)
```

```python
business.shape
```

```python
business.head()
```

After choosing categories, we will now just focus on the state with most businesses available!

```python
temp = business['state'].value_counts()
```

```python
temp.index[:1].tolist()
```

```python
business = business[business['state'].isin(temp.index[:1].tolist())]
business.shape
```

Let's just narrow down even further by focusing on business with review counts at least 10 because businesses with less than 10 reviews may have the good review comming from its owner and all the bad reviews comming from its competitors

```python
business = business[business['review_count'] >= 10]
business.shape
```

```python
business.head()
```

```python
business['attributes.HairSpecializesIn.coloring'].unique()
```

```python
business = business[business['attributes.HairSpecializesIn.coloring'].isnull()]
business.shape
```

```python
business.columns
```

```python
business.dropna(axis = 1, how= "all", inplace = True)
```

```python
business.head()
```

```python
business.columns
```

```python
def get_attribute(business):
    attributes = []
    for i in business.index:
        attribute = []
        for c in business.columns:
            if c == "attributes.RestaurantsPriceRange2":
                c = "PriceRange"
            elif 'attributes' in c and business.loc[i][c] == True:
                    attr = "_".join(c.replace("attributes.", "").split("."))
                    attribute.append(attr)
        attributes.append(attribute)
    return attributes
```

```python
start = time.time()
attributes = get_attribute(business)
end = time.time()
print(end - start)
```

```python
attributes
```

```python
business['attributes'] = attributes
```

```python
for c in business.columns:
    if 'attributes.' in c:
        business.drop(c, axis = 1, inplace = True)
```

```python
business.head()
```

```python
business.columns
```

```python
business.shape
```

# Clean User and Review

```python
user = pd.read_csv('user.csv')
```

```python
user.shape
```

Because of the cold-start problem, we will just use user with at least 3 reviews

```python
user = user[user['review_count'] > 2]
user.shape
```

```python
review = pd.read_csv('review.csv')
```

```python
review.shape
```

```python
review = review[review['business_id'].isin(business['business_id'].values.tolist())]
```

```python
review.shape
```

```python
review.columns
```

```python
updated_review_count = review['user_id'].value_counts()
```

```python
updated_review_count
```

```python
user.columns
```

```python
user = user[user['user_id'].isin(review['user_id'].tolist())]
```

```python
user.shape
```

```python
user['updated_review_count'] = user['user_id'].map(updated_review_count)
```

```python
user.columns
```

```python
user.drop(['review_count'],axis = 1, inplace = True)
```

```python
user.rename(columns = {'updated_review_count': 'review_count'},inplace = True)
user.columns
```

```python
user.shape
```

```python
review.shape
```

```python
user = user[user['review_count'] > 2]
```

```python
review = review[review['user_id'].isin(user['user_id'].values.tolist())]
```

```python
print(user.shape)
print(review.shape)
```

```python
user.to_csv('user_cleaned.csv', encoding = 'utf-8', index = False)
```

```python
review.to_csv('review_cleaned.csv', encoding = 'utf-8', index = False)
```
