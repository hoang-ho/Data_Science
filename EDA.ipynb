# EDA

```python
import pandas as pd
from ast import literal_eval
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
matplotlib.style.use("ggplot")
import seaborn as sns
```

```python
business = pd.read_csv('business_cleaned(1).csv')
```

```python
business
```

```python
business['categories'] = [list(filter(None, literal_eval(x))) for x in business.categories]
```

```python
business['categories']
```

```python
plt.figure(figsize = (10,5))
plt.hist(business['stars'], bins = 8, color = 'b')
plt.xlabel('Business Average Ratings')
plt.ylabel('Count')
plt.title('Distribution of Business Average Rating')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.scatter(business['stars'], business['review_count'], color = 'b')
plt.xlabel('Business Average Ratings')
plt.ylabel('Review Count')
plt.title('Review Count and Business Average Ratings')
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.hist(user['average_stars'], bins = 8, color = 'b')
plt.xlabel('User rating')
plt.ylabel('Count')
plt.title('Distribution of Users Rating')
plt.show()
```
